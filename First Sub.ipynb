```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression
```

```python
training_data = pd.read_csv("file:///C:/Users/DELL/Downloads/Train_Data.csv")
testing_data = pd.read_csv("file:///C:/Users/DELL/Downloads/Test_Data.csv")
```

```python
X = training_data.drop(["campaign","revenue","date","adgroup","ad","cost"],axis=1)
y = training_data["revenue"]
X
```

```python
linreg = LinearRegression()
linreg.fit(X,y)
```

```python
testing_data.drop(["campaign","date","adgroup","ad","cost"],axis=1,inplace=True)
```

```python
arr = linreg.predict(testing_data)
print(arr)
for i in range(len(arr)):
    if arr[i]<0:
        arr[i]=0 
```

```python
res = pd.DataFrame(arr)
res.index = testing_data.index
res.columns = ["revenue"]
res.to_csv("first_sub.csv", index = False)   
```
