```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt 
from sklearn.metrics import mean_squared_error
from sklearn.metrics import accuracy_score,r2_score
from sklearn.linear_model import Ridge
```

```python
data = pd.read_csv("file:///C:/Users/DELL/Downloads/Train_Data.csv")
data.head()
```

```python
# data.drop(["campaign"],axis=1,inplace=True)
data.head()
```

```python
beta = data[data.revenue !=0 ]
beta
```

```python
yyy = []
xxx = []

for index in beta.index:
    if "-02-" in beta.date[index]:
        yyy.append(index)
    else:
        xxx.append(index)
Xeta = beta.loc[xxx,:]
yeta = beta.loc[yyy,:]

#checker for 0 values
if 0 in Xeta.cost.unique():
    print("YES")
else:
    print("NO")
```

```python
Xeta.impressions = np.log(Xeta.impressions)
Xeta.clicks = np.log(Xeta.clicks)
Xeta.conversions = np.log(Xeta.conversions)
Xeta.cost = np.log(Xeta.cost)
Xeta.revenue = np.log(Xeta.revenue)
Xeta
```

```python
# Xeta.drop(["date","adgroup","ad"],axis=1,inplace=True)
# yeta.drop(["date","adgroup","ad"],axis=1,inplace=True)
```

```python
y_Xeta = Xeta["revenue"]
X_Xeta = Xeta.drop("revenue",axis=1)
```

```python
X_Xeta
```

```python
rid_reg = Ridge(alpha=2.77,random_state=781)
```

```python
rid_reg.fit(X_Xeta,y_Xeta)
```

```python
#transforming testing data
yeta.impressions = np.log(yeta.impressions)
yeta.clicks = np.log(yeta.clicks)
yeta.conversions = np.log(yeta.conversions)
yeta.cost = np.log(yeta.cost)
yeta
```

```python
y_test = yeta["revenue"]
X_test = yeta.drop("revenue",axis=1)
```

```python
ans = rid_reg.predict(X_test)
```

```python
print(np.sqrt(mean_squared_error(np.exp(ans),y_test)))
```
